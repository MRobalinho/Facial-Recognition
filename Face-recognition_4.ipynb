```python
# Import packages
import face_recognition
import cv2
import win32com.client as wincl
import pandas as pd
```

```python
# -- Images to Learn ------------
first_path = 'images'
path = 'images' + '/'
#----------------------------------------
```

```python
# Read index file (excel file)

index_file = path + 'images_index.xls'
df = pd.read_excel(index_file)
df.columns = ["image", "name"]
df = pd.DataFrame(df)
# See imported excel table
df
```

```python
# Transform dataframe to a list
index_list = df.values.tolist()
index_list
```

```python
# read Images from folder images
import glob
xfiles = glob.glob(path +"*.jpg")
print(xfiles)
```

```python
# Some initiations for video record
PADDING = 50
ready_to_detect_identity = True
windows10_voice_interface = wincl.Dispatch("SAPI.SpVoice")

global ready_to_detect_identity
global old_name
```

```python
# Output video record
filename="output\VideoCap1.avi" #[place were i stored my output file]

codec = cv2.VideoWriter_fourcc(*'MJPG')  # AVI
#codec = cv2.VideoWriter_fourcc('m','p','4','v')#fourcc stands for four character code
framerate=2
resolution=(640,480)
    
VideoFileOutput=cv2.VideoWriter(filename,codec,framerate, resolution)
```

```python
VideoFileOutput
```

```python
def welcome_users(identities):
    """ Outputs a welcome audio message to the users """
    
    welcome_message = 'Welcome ,'  + identities +', have a nice day.' 
            
    windows10_voice_interface.Speak(welcome_message)

    # Allow the program to start detecting identities again
    ready_to_detect_identity = True
```

```python
# Some tests
new_path = first_path + '\\'
x_image = 'images\\manuel.jpg'
x_image = x_image.replace(new_path, '')
x_image 
```

```python
# Create a list with the names
x_known_face_names = []

print('---- Processing Images - Names ---')
x = 0
list_len = len(xfiles)

while( x < list_len):
#    print('Reading names:', x, xfiles[x])
    x_image = xfiles[x]
    # Extract name path to the name file
    x_image = x_image.replace(new_path, '')
    y_name = df.loc[df.image == x_image]
 
    if (y_name.empty == True):
        print ( 'Not Found:', x_image, ' Name:' ,y_name.name)
    else:
        print ( 'Found:', x_image, ' Name:', y_name.name)
        xx = y_name['name'].values.tolist()
        x_known_face_names.append(xx)    # Create a list
    x+=1
    
x_known_face_names  
```

```python
# Load a sample picture and learn how to recognize it.

x_face_encoding = []

print('---- Processing Images - Learning ---')
x = 0
list_len = len(xfiles)
while( x < list_len):
    print('Reading:', x, xfiles[x])
 #   x_image.append(xfiles[x])
    x_image = xfiles[x]
    x_image_face = face_recognition.load_image_file(x_image)
    x_face_encoding.append(face_recognition.face_encodings(x_image_face)[0])
    x+=1

print('--------------------------')
print('---- FACE ENCODING ---')
print(x_face_encoding)
```

```python
# Get a reference to webcam #0 (the default one)
video_capture = cv2.VideoCapture(0)

# -------------- LEARNED IMAGES ---------------------------

# Create arrays of known face encodings and their names
known_face_encodings = x_face_encoding

# Names of the images
# Transform lists to Dataframe
df_name = pd.DataFrame(list(x_known_face_names))
df_name.columns = ['name']

# Transform DataFrame to List
known_face_names = list(df_name.name)
known_face_names
```

```python
# TRY RECOGNITION IN VIDEO CAMERA
old_name = "Unknown"
while True:
    # Grab a single frame of video
    ret, frame = video_capture.read()

    # Convert the image from BGR color (which OpenCV uses) to RGB color (which face_recognition uses)
    rgb_frame = frame[:, :, ::-1]

    # Find all the faces and face enqcodings in the frame of video
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)

    # Loop through each face in this frame of video
    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        # See if the face is a match for the known face(s)
        matches = face_recognition.compare_faces(known_face_encodings, face_encoding)

        find_name = "Unknown"

        # If a match was found in known_face_encodings, just use the first one.
        if True in matches:
            first_match_index = matches.index(True)
            find_name = known_face_names[first_match_index]
            
            if find_name != old_name:
                old_name = find_name
                welcome_users(find_name)  # Speak my name
              
            
        # Draw a box around the face
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)

        # Draw a label with a name below the face
        cv2.rectangle(frame, (left, bottom - 35), (right, bottom), (0, 0, 255), cv2.FILLED)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, find_name, (left + 6, bottom - 6), font, 1.0, (255, 255, 255), 1)
    
    # Write at output video
    VideoFileOutput.write(frame)
    
    # Display the resulting image
    cv2.imshow('Video preview-MRobalinho', frame)
        
    # Hit 'q' on the keyboard to quit!
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release handle to the webcam
# Output Video
print('Key q Pressed - Exit')
VideoFileOutput.release()
#
video_capture.release()
cv2.destroyAllWindows()
```
